# **Feature Selection with ExtraTress Classifier**
* *Feature selection* has to do with the selection of features or attributes in a dataset that are most important for predicting the target variable. In other words, which features determine the output (target variable). There are several ways to do this. In this notebook we will use EXtraTrees Classifier to make this selection.

* *ExtraTrees Classifier* is clustering tree-based algorithm. Tree-based algorithms are considered to be very robust, making this a good option.

* he algorithm creates branches from features in the entire dataset at random. It is an ensemble model.

Let's see how this works.

But just before we do this let's talk about the dataset.

**UNSW-NB15 Dataset**


*   This is a dataset containing network data generated by the IXIA PerfectStorm tool of the Australian Centre for Cyber Security (ACCS).
*   It contains 9 attack categories inclusive of Fuzzers, DDoS, Worms and Backdoors
* This data was processed then saved as a pickle on Google drive.



```python
# import required libraries
import pandas as pd
from google.colab import files
```

```python
# import cleaned dataset
dataset = pd.read_pickle("/content/drive/MyDrive/network dataset/df-bulk2.pk1")
```

```python
X = dataset.drop('label', axis=1) # features
y = dataset['label'] # labels
```

```python
from sklearn.ensemble import ExtraTreesClassifier

# Feature Importance with Extra Trees Classifier

X2 = X.values #features from dataset
Y2 = y.values #labels from dataset

fs2 = [] #create a list to hold selected features
n_estimators = 50
# feature extraction step
model = ExtraTreesClassifier(n_estimators=n_estimators) #get n_estimator value from oob curve calculation
model.fit(X2, Y2)
fset = list(model.feature_importances_>0.015) # add features to a list based on importance scores
i = range(len(fset))
for item in i:
  if fset[item] == True:
    fs2.append(item)  #get column indexes
#print(fs2)
#create dataframes from list items
fs2_X = X.iloc[:, fs2] #create a subset of the data containing only the selected features
print(len(fs2), " features were selected")
```

**Note:**
19 features were selected based on their feature importance values being greater than 0.015.

```python
# Print a list of the column names of the selected features
print(list(fs2_X.columns))
```

```python
#save this list
fs2_X.to_pickle("/content/drive/MyDrive/Pickles/et_selx.pk1")
```
